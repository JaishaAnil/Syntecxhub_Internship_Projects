```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
plt.style.use('ggplot')
```

```python
df = pd.read_csv("WEBSITE-TRAFFIC-ANALYSIS.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df["event_name"].value_counts()
```

```python
plt.figure(figsize=(10,5))
df["event_name"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Website Events")
plt.xlabel("Event Name")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df["traffic_source.source"].value_counts()
```

```python
plt.figure(figsize=(8,5))
df["traffic_source.source"].value_counts().head(10).plot(kind="bar", color="skyblue")
plt.title("Top Traffic Sources")
plt.xlabel("Traffic Source")
plt.ylabel("Number of Events")
plt.xticks(rotation=45)
plt.show()
```

```python
df["device.category"].value_counts()
```

```python
plt.figure(figsize=(6,6))
df["device.category"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.ylabel("")
plt.title("Device Category Distribution")
plt.show()
```

```python
df["geo.country"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
df["geo.country"].value_counts().head(10).plot(kind="bar", color="green")
plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Events")
plt.xticks(rotation=45)
plt.show()
```

```python
df["device.web_info.browser"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
df["device.web_info.browser"].value_counts().head(10).plot(kind="bar", color="orange")
plt.title("Top Browsers")
plt.xlabel("Browser")
plt.ylabel("Number of Events")
plt.xticks(rotation=45)
plt.show()
```

```python
df["ecommerce.purchase_revenue"].describe()
```

```python
plt.figure(figsize=(8,5))
df["ecommerce.purchase_revenue"].dropna().hist(bins=20)
plt.title("Purchase Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

## Business Insights

- The most frequent website event was **page_view**, indicating that users primarily browsed the website.
- The majority of traffic originated from the top traffic sources identified in the analysis.
- Most users accessed the website using **mobile devices**, showing the importance of mobile optimization.
- A few countries contributed the majority of website traffic.
- Purchase events were fewer than browsing events, suggesting an opportunity to improve conversion rates.

## Conclusion

This project analyzed website traffic using a Google Analytics 4 dataset. The analysis explored user events, traffic sources, device categories, geographic distribution, browsers, and purchase revenue. Various visualizations were created to identify user behavior and website performance. The findings can help improve marketing strategies, user engagement, and overall website performance.
